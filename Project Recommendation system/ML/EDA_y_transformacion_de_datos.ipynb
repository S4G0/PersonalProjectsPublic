##**3. Transformación de datos**

###**3.1. Transformación de datos de productos**

```python
#Me quedo solo con las variables que necesito
df_products_transformed = df_products[["product_id", 	"name", "category", "descripcion", "palabras_clave"]]

df_products_transformed["palabras_clave"] = df_products_transformed["palabras_clave"].apply(lambda x: ", ".join(sorted(x.split(", "))))

# Hacer groupby en 'categoria' y 'subcategoria' y resetear el índice
df_grouped = df_products_transformed.groupby(["name", "category", "descripcion", "palabras_clave"]).sum().reset_index().drop(['product_id'], axis=1)

# Asignar un índice único
df_grouped['index'] = range(1, len(df_grouped) + 1)

df_products_transformed = pd.merge(df_products_transformed[["product_id", 	"name"]], df_grouped, on='name', how='left')
```

```python
df_products_transformed
```

###**3.2. Transformación de datos de usuarios**

```python
#Me quedo solo con las variables que necesito
df_users_transformed = df_users[["user_id", "edad",	"genero", "intereses"]]

# Define bins starting at 18 with intervals of 12
bins = list(range(18, df_users_transformed['edad'].max() + 12, 12))

# Create labels for categories
labels = [f"{bins[i]}-{bins[i+1]-1}" for i in range(len(bins)-1)]

# Apply binning
df_users_transformed['edad_category'] = pd.cut(df_users_transformed['edad'], bins=bins, labels=labels, right=False)

df_users_transformed.drop(['edad'], axis=1, inplace=True)
```

```python
df_users_transformed
```

###**3.3. Transformación de datos de interacciones**

```python
#Me quedo solo con las variables que necesito
df_interactions_transformed = df_interactions[["user_id",	"product_id",	"tipo_interaccion", "rating",	"comentario"]]
#Elimino las filas que no tienen ni calificación ni comentario
df_interactions_transformed = df_interactions_transformed[(df_interactions_transformed["comentario"].notna())|(df_interactions_transformed["rating"].notna())]
#Creando columna que une comentarios y rating
df_interactions_transformed['rating_y_comentarios'] = df_interactions_transformed['rating'].fillna(df_interactions_transformed['comentario'])
#Quitando variables de rating y comentario
df_interactions_transformed.drop(['rating', 'comentario'], axis=1, inplace=True)

# Diccionario de mapeo
map_dict = {
    'Excelente': 5,
    'Recomendado': 5,
    'Podría mejorar': 3,
    'Aceptable': 3,
    'Malo': 1
}

# Crear nueva columna transformada
df_interactions_transformed['nuevo_rating'] = df_interactions_transformed['rating_y_comentarios'].map(map_dict).fillna(df_interactions_transformed['rating_y_comentarios']).astype(int)
# Remover columna que une comentarios y rating
df_interactions_transformed = df_interactions_transformed.drop(['rating_y_comentarios'], axis=1).reset_index(drop=True)
# Escalando columna con max-min
df_interactions_transformed['nuevo_rating'] = (df_interactions_transformed['nuevo_rating'] - 1) / (5 - 1)

df_interactions_transformed = pd.merge(df_interactions_transformed, df_products_transformed[['product_id','index']], on='product_id', how='left')
```
